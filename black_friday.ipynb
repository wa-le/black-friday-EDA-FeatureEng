```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
#import test dataset
df_test = pd.read_csv("/content/drive/MyDrive/Data/Black-Friday/test.csv")
df_test
```

```python
# concatenate both dataset
df_train
df_test
```

```python
df_both = pd.concat([df_train, df_test])
df_both.reset_index()
```

```python
df = df_both.copy()
```

```python
df.describe()
```

```python
# drop useless cols
df.drop(["User_ID"], axis=1, inplace=True)
```

```python
df.head(2)
```

```python
# Handling Categorical feature Gender
# replace Gender with 0 and 1
df['Gender'].replace({'F':0, 'M':1}, inplace=True)
```

```python
df.info()
```

```python
# Handling Categorical feature Age
df['Age'].unique()
```

```python
df['Age'].replace({'0-17':1, '18-25':2, '26-35':3, '36-45': 4, '46-50':5, '51-55':6, '55+':7 }, inplace=True)
```

```python
df
```

```python
# data correct till here
df1 = df.copy()
# In case data gets messed up, come back here
# My Checkpoint
```

```python
df1.info()
```

```python
# fixing categorical city_category
df1_city = pd.get_dummies(df1['City_Category'], drop_first=True)
# having 3 categories, 2 categories are sufficient to represent the 3
```

```python
# now combine df1_city with df1 - concat
df1 = pd.concat([df1, df1_city], axis=1)
df1
```

```python
# drop the column
df1.drop(['City_Category'], axis=1, inplace=True)
df1
# "B", "C" now represents the City Category column which has now been dropped in the latest DF(df1)
```
